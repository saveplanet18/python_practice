# Switch Case in Python (Replacement)
## dictionary mapping()

```python
def show(args):
    switch = {
        0 : 'zero',
        1 : 'one',
        2 : 'two',
    }
    return switch .get(args)
```

```python
show(1)
```

```python
if __name__=="__main__":
        args = 0
        print(show(args))
```

# Generators in Python

```python
def sumfun():
    yield 1
    yield 2
    yield 3
for i in sumfun():
    for j in range(5):
        print("inner loop", j)
    print("outer loop :", i)
```

```python
x = sumfun()
```

```python
x.next()
```

```python
for i in x:
    print(i)
```

```python
for j in x:
    print(j,x)
```

```python
def gene():
    t = 1
    print("first result is ",t)
    yield t
```

```python
gene()
```

```python
call = gene()
```

```python
next(call)
```

```python
def ne():
    t = 1
    print("first element: ",t)
    yield t
    t+=1
    print("the second element: ",t)
    yield t
    t+=2
    print("third element: ",t)
    yield t
```

```python
call = ne()
```

```python
try:
    for x in call:
        print("all ",x)
except:
    print("pass")
```
