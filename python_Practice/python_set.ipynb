```python
for i in range(0,5):
    for j in range(0,i+1):
        print("*",end="")
    print()
```

```python
def show(n):
    for i in range(0,n):
        for j in range(0,i+1):
            print("*",end='')
        print()
```

```python
show(5)
```

```python
def show(n):
    mylist = []
    for i in range(1,n+1):
        mylist.append("*"*i)
    print("\n".join(mylist))
```

```python
show(5)
```
