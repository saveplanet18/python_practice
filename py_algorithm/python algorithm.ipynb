```python
#linear search
def search(arr,n):
    for index , values in enumerate(arr):
        if values==n:
            return index
    return -1
arr=[10,20,30,45,80,90]
n=30
print(n,"n is presented in values ",search(arr,n))
```

```python
def sum1(n):
    """
    take an input of n and return the sume
    """
    final_sum=0
    for x in range(n+1):
        final_sum+=x
    return final_sum
```

```python
sum1(5)
```

```python
from math import log
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')
# set up runtime comparision
n=np.linspace(1,10)
labels=['Constant','logarithmic','linear','log liner','Quadratic','Cubic','Expontial']
big_o=[np.ones(n.shape),np.log(n),n*np.log(n),n**2,n**3,2**n]

plt.figure(figsize=(12,10))
plt.ylim(0,50)

for i in range(len(big_o)):
    plt.plot(n,big_o[i],label=labels[i])
    
plt.legend(loc=0)
plt.ylabel('relative runtime')
plt.xlabel('n')
```
